```python
# ratings
import ast
ted.ratings = ted.ratings.apply(ast.literal_eval)
```

```python
# total number of ratings received by each talk
def ratings_per_talk(ratings):
    num = 0
    for i in ratings:
        num += i['count']
    return num
ted['ratings_per_talk'] = ted.ratings.apply(ratings_per_talk)
```

```python
def funny_count(ratings):
    for i in ratings:
        if i['name'] == 'Funny':
            return i['count']

ted['Funny_Index'] = ted.ratings.apply(funny_count)
```

```python
ted['Funny_Rate'] = ted.Funny_Index / ted.ratings_per_talk
```

```python
ted.sort_values('Funny_Rate').speaker_occupation.tail(20)
```
